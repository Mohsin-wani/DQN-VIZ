```python
## the algorithm chosen has to be imported, three options provided are
## DoubleDQN, DuelingDQN and DQN
## Refer to readme file for setting parameters
## for Code ocean capsule we are only recording videos to results folder

import DoubleDQN as ddqn

# instantiate environment by passing the environment name gym/atari
print(ddqn.os.getcwd())
environment = ddqn.Environment('CartPole-v0')
env = environment.get_env()
#set the hyperparameters, e.g. length of training, number of episodes, recording params etc.
hyp = ddqn.Hyperparameters(env, record=True)

#pass the environment and hyperparameters to the agent and trains.
agent = ddqn.Agent(env,hyp)

agent.dqn()
```
